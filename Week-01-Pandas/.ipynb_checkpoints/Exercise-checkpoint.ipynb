## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 
condition = df.price < 100
len(df[condition])
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.
df_bk = df['neighbourhood_group'] == 'Brooklyn'
len(df[df_bk])
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.

# selecting Brooklyn
condition_1 = df['neighbourhood_group'] == 'Brooklyn'

# select Prices that are less than 100
condition_2 = df['price'] < 100

# select all data that meets both requirments
len(df[condition_1 & condition_2])
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.

# selecting list of names
host_names = ['Michael', 'David', 'John', 'Daniel']

# creating selection condition 
condition = df.host_name.isin(host_names)

# display those who have the host name of ...
df[condition]
df[condition]['host_name'].value_counts()
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  

c1 = df['neighbourhood'] == 'Williamsburg'

# set equal to 100 when both conditions are true, and zero when not true.
df['williamsburg_fee'] = np.where(c1, 100, 0)
df['adjusted_price'] = df['price'] + df['williamsburg_fee']
df[c1].head(2)

# or
# c1 = df['neighbourhood'] == 'Williamsburg'
# df['adjusted_price'] = np.where(c1, df['price'] + 100, df['price'])
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

df_private_rooms = df.room_type.value_counts() / df.room_type.value_counts().sum()
df_private_rooms

# or
# private_rooms = df['room_type'] == "Private room"
# shared_rooms =  df['room_type'] == "Shared room"

# private_percent = len(df[private_rooms]) / len(df)
# shared_percent = len(df[shared_rooms]) / len(df)

# print(private_percent)
# print(shared_percent)
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.
# df.groupby('neighbourhood_group').count()

listings = df.groupby('neighbourhood_group')
listings['neighbourhood_group'].count()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
df.groupby('neighbourhood_group')['price'].mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
gb = df.groupby('neighbourhood_group')
gb_stats = gb['price'].agg( ['min', 'max'] )
gb_stats
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.

gb = df.groupby(['neighbourhood_group','room_type'])
gb['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
grouped_cols = ['neighbourhood_group', 'room_type']
gb = df.groupby(grouped_cols)
gb['price'].agg(['min', 'max', 'mean', 'median', 'std'])

#df.groupby(['neighbourhood_group','room_type'])['price'].agg( ['count', 'min', 'max', 'mean', 'median', 'std', 'sum'] )
```

```python
# 2. Who won Best Rap Album in 2009?

best_rap_condition = df.category == "Best Rap Album"
year_condition = df.year == 2009
winner_condition = df.winner == True
df[best_rap_condition & year_condition & winner_condition]
```
